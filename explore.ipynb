## 2. Explore your Datasets

Now that you have your list objects created and cleaned, let's explore the descriptive statistics of these two files and make subsequent observations on our data.

```python
# TODO: Calculate the standard deviation of both the control and treatment group
std_control = stats.stdev(clean_control)
std_treatment = stats.stdev(clean_treatment)

# TODO: print your findings
print(std_control)
print(std_treatment)
```

```python
# TODO: Calculate the mean of both the control and treatment group
mean_control = stats.mean(clean_control)
mean_treatment = stats.mean(clean_treatment)

# TODO: print your findings
print(mean_control)
print(mean_treatment)
```

```python
# median 
med_control = stats.median(clean_control)
med_treatment = stats.median(clean_treatment) 

print(med_control)
print(med_treatment)
```

```python
range_control = max(clean_control) - min(clean_control)
range_treatment = max(clean_treatment) - min(clean_treatment)

print(range_control)
print(range_treatment)
```

```python
# TODO: Are there any outliers in your control group? (use 2 * stdev to find these outliers)
outliers_control = []
upper_lim_control = mean_control + (2 * std_control)
lower_lim_control = mean_control - (2 * std_control)
for item in clean_control:
    if item > upper_lim_control or item < lower_lim_control:
        outliers_control.append(item)

print(outliers_control)
```

```python
# TODO: Are there any outliers in your treatment group?
outliers_treatment = []
upper_lim_treat = mean_treatment + (2 * std_treatment)
lower_lim_treat = mean_treatment - (2 * std_treatment)
for item in clean_treatment:
    if item > upper_lim_treat or item < lower_lim_treat:
        outliers_treatment.append(item)

print(outliers_treatment)
```
